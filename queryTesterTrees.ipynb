```python
import ydf
import numpy as np
import pandas as pd

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Load the decision tree model
tree_model = ydf.load_model("/content/drive/MyDrive/Colab Notebooks/decision_tree_model/")

# Load the random forest model
forest_model = ydf.load_model("/content/drive/MyDrive/Colab Notebooks/random_forest_model/")

# Read data from the text file
with open("/content/drive/MyDrive/Colab Notebooks/hsv_values.txt", "r") as file:
    lines = file.readlines()
    print(f"info:{lines}")
# Skip the first lines (column names and non-numeric identifiers)
lines = lines[1:]

# Convert the remaining data into a NumPy array and adjust data type if needed
data = []
names = []
hex_values = []
for line in lines:
    row = line.strip().split(',')
    names.append(row[0])  # Save the name
    hex_values.append(row[1])  # Save the hex value
    data.append(row[2:])  # Skip the first two columns (Name and Hex)
query_data = np.array(data, dtype=np.float32)

# Convert the NumPy array to a Pandas DataFrame
# Adjust the column names accordingly
column_names = ['Red (8 bit)', 'Green (8 bit)', 'Blue (8 bit)',
                'Hue (degrees)', 'HSL.S (%)', 'HSL.L (%) HSV.S (%) HSV.V (%)']
query_df = pd.DataFrame(query_data, columns=column_names)

# Add the names and hex values as separate columns
query_df.insert(0, 'Name', names)
query_df.insert(1, 'Hex (24 bit)', hex_values)

# Make predictions using the decision tree model
tree_predictions = tree_model.predict(query_df)

# Make predictions using the random forest model
forest_predictions = forest_model.predict(query_df)

# Define the class labels
class_labels = ['Red', 'Green', 'Blue', 'Yellow', 'Orange', 'Purple', 'Neutral']
# 'White','Black','Gray',
# Convert the decision tree predictions to class labels
tree_predictions_labels = [class_labels[i] for i in np.argmax(tree_predictions, axis=1)]

# Convert the random forest predictions to class labels
forest_predictions_labels = [class_labels[i] for i in np.argmax(forest_predictions, axis=1)]
# Make predictions using the decision tree model
tree_predictions = tree_model.predict(query_df)
print("Raw Decision Tree Predictions:", tree_predictions)

# Make predictions using the random forest model
forest_predictions = forest_model.predict(query_df)
print("Raw Random Forest Predictions:", forest_predictions)

# Print predictions
print("Decision Tree Predictions:", tree_predictions_labels)
print("Random Forest Predictions:", forest_predictions_labels)
```
